```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from nbodykit import cosmology
from nbodykit.cosmology import Planck15
```

```python
redshift   = 0.0
cosmo      = Planck15
EHPower    = cosmology.LinearPower(cosmo,
                                   redshift=redshift,
                                   transfer='EisensteinHu')
nowEHPower = cosmology.LinearPower(cosmo,
                                   redshift=redshift,
                                   transfer='NoWiggleEisensteinHu')
```

```python
k = np.linspace(1.e-3, 10, num=10000)
pkwig = EHPower(k)
pknow = nowEHPower(k)
```

```python
plt.plot(k, pkwig/pknow)
plt.ylabel('Pwig/Pnowig')
plt.xlabel('k')
plt.xscale('log')
```
